```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

city_data_df.head()
ride_data_df.head()

# Combine the data into a single dataset
combo_df = pd.DataFrame()
combo_df = pd.merge(city_data_df, ride_data_df, on='city')

# Display the data table for preview
combo_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

ride_share_data_df = pd.DataFrame()
ride_share_data_df = combo_df[['city', 'type', 'fare']]

ride_share_data_df = ride_share_data_df.groupby(['city', 'type']).fare.agg(['sum', 'count', 'mean'])
ride_share_data_df = ride_share_data_df[['count', 'mean', 'sum']]

urban_ride_share_df = pd.DataFrame()
urban_ride_share_df = ride_share_data_df.filter(like='Urban', axis=0)
urban_count_x = urban_ride_share_df['count'].values.tolist()
urban_fare_y = urban_ride_share_df['mean'].values.tolist()

suburban_ride_share_df = pd.DataFrame()
suburban_ride_share_df = ride_share_data_df.filter(like='Suburban', axis=0)
suburban_count_x = suburban_ride_share_df['count'].values.tolist()
suburban_fare_y = suburban_ride_share_df['mean'].values.tolist()

rural_ride_share_df = pd.DataFrame()
rural_ride_share_df = ride_share_data_df.filter(like='Rural', axis=0)
rural_count_x = rural_ride_share_df['count'].values.tolist()
rural_fare_y = rural_ride_share_df['mean'].values.tolist()

# Build the scatter plots for each city types
# Incorporate a text label regarding circle size
# Save Figure

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

a1 = np.array( urban_count_x )
a2 = np.array( suburban_count_x )
a3 = np.array( rural_count_x )
area1 = np.pi * (0.4 * a1)**2 
area2 = np.pi * (0.4 * a2)**2 
area3 = np.pi * (0.4 * a3)**2 

plots= [(urban_ride_share_df, 'orange', 'Urban', area1),
       (suburban_ride_share_df, 'blue', 'Suburban', area2),
       (rural_ride_share_df, 'yellow', 'Rural', area3)]

for df, color, label, area in plots:       
    df.plot.scatter(ax=ax, x='count', y='mean',color=color, label=label, s=area, edgecolors="black",alpha=0.5)

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
ax.set_title("Pyber Ride Sharing Data (2016)")

ax.legend(title='City Type')

ax.grid(True)
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
ride_share_data_df = pd.DataFrame()
ride_share_data_df = combo_df[['city', 'type', 'fare']]
s = ride_share_data_df['fare']
total_fare = s.sum()

ride_share_data_df = ride_share_data_df.groupby(['type']).fare.agg(['sum', 'count', 'mean'])
ride_share_data_df = ride_share_data_df[['count', 'mean', 'sum']]

urban_ride_share_df = ride_share_data_df.filter(like='Urban', axis=0)
suburban_ride_share_df = ride_share_data_df.filter(like='Suburban', axis=0)
rural_ride_share_df = ride_share_data_df.filter(like='Rural', axis=0)

pct_urban = urban_ride_share_df['sum'] / total_fare
pct_suburban = suburban_ride_share_df['sum'] / total_fare
pct_rural = rural_ride_share_df['sum'] / total_fare

# Build Pie Chart
# Save Figure
labels ='Urban',  'Rural', 'Suburban'
sizes = [pct_urban, pct_rural, pct_suburban]
colors = ['lightcoral', 'yellow', 'lightskyblue']
explode = (0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=-75)

ax1.set_title("% of Total Fares by City Type")
# Save Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

ride_share_data_df = pd.DataFrame()
ride_share_data_df = combo_df[['city', 'type', 'fare']]

ride_share_data_df = ride_share_data_df.groupby(['type']).fare.agg(['sum', 'count', 'mean'])
ride_share_data_df = ride_share_data_df[['count', 'mean', 'sum']]

s = ride_share_data_df['count']
total_rides = s.sum()

urban_ride_share_df = ride_share_data_df.filter(like='Urban', axis=0)
suburban_ride_share_df = ride_share_data_df.filter(like='Suburban', axis=0)
rural_ride_share_df = ride_share_data_df.filter(like='Rural', axis=0)

pct_urban_rides = urban_ride_share_df['count'] / total_rides
pct_suburban_rides = suburban_ride_share_df['count'] / total_rides
pct_rural_rides = rural_ride_share_df['count'] / total_rides

# Build Pie Chart
# Save Figure

labels ='Urban',  'Rural', 'Suburban'
sizes = [pct_urban_rides, pct_rural_rides, pct_suburban_rides]
colors = ['lightcoral', 'yellow', 'lightskyblue']
explode = (0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=-75)

ax1.set_title("% of Total Rides by City Type")

# Show Figure
plt.show()

```
