```python
import pymysql
import pandas as pd
import csv
```

```python
conn = pymysql.connect(host='localhost',
                            user='root',
                            port=3306,
                            password='admin',
                            db='chargingS_dealers')
```

```python
mycursor = conn.cursor()
dataSt = []
Statdf = []
mycursor.execute("SELECT d.busname, d.lotaddr, d.phone,d.lotcity,d.lotstate, s.station_name, s.station_phone, s.street_address, s.city,s.state FROM chargs_dealers d JOIN charging_stations s ON UPPER(trim(d.lotcity)) = UPPER(trim(s.city)) and UPPER(trim(d.lotstate)) = UPPER(trim(s.state))")
#mycursor.execute("SELECT d.busname, d.lotaddr, d.phone,d.lotcity,d.lotstate, s.station_name, s.station_phone, s.street_address, s.city,s.state FROM chargs_dealers d INNER JOIN charging_stations s ON d.lotcity = s.city and d.lotstate = s.state group by s.station_name")
myresult = mycursor.fetchall()
dataSt = {'Dealername': [], 'Dealeraddr': [], 'DealerPhone': [],'DealerCity':[],'DealerState':[],'StationName': [],'StationPhone': [], 'StreetAddress': [],'StationCity':[],'StationState':[]}
for statDt2 in myresult:
    #print(statDt2[3])
    dataSt['Dealername'].append(statDt2[0])
    dataSt['Dealeraddr'].append(statDt2[1])
    dataSt['DealerPhone'].append(statDt2[2])
    dataSt['DealerCity'].append(statDt2[3])
    dataSt['DealerState'].append(statDt2[4])
    dataSt['StationName'].append(statDt2[5])
    dataSt['StationPhone'].append(statDt2[6].replace("None", "999-999-9999"))
    dataSt['StreetAddress'].append(statDt2[7])
    dataSt['StationCity'].append(statDt2[8])
    dataSt['StationState'].append(statDt2[9])
                                  
Statdf = pd.DataFrame(dataSt)
    
Statdf.head(10)
```

```python
Statdf.to_csv(r'Index.csv')
```
